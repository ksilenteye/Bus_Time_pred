```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import LabelEncoder
import numpy as np
from datetime import datetime
```

```python
# Convert 'Sch_Trip_Start' and 'Actual_Trip_End' to datetime
DS['Sch_Trip_Start'] = pd.to_datetime(DS['Sch_Trip_Start'], format='%d-%m-%Y %H:%M')
DS['Actual_Trip_End'] = pd.to_datetime(DS['Actual_Trip_End'], format='%d-%m-%Y %H:%M')

# Extract features from the datetime columns
DS['Sch_Start_Hour'] = DS['Sch_Trip_Start'].dt.hour
DS['Sch_Start_Minute'] = DS['Sch_Trip_Start'].dt.minute
```

```python
# Encode 'Operated_Day' and 'Route_Name' into numerical values
label_encoder_day = LabelEncoder()
label_encoder_route = LabelEncoder()

DS['Operated_Day_Encoded'] = label_encoder_day.fit_transform(DS['Operated_Day'])
DS['Route_Name_Encoded'] = label_encoder_route.fit_transform(DS['Route_Name'])

# Calculate trip duration in minutes (target variable)
DS['Actual_Trip_End_Minutes'] = (DS['Actual_Trip_End'] - DS['Sch_Trip_Start']).dt.total_seconds() / 60.0
```

```python
plt.subplot(2, 2, 1)
sns.histplot(DS['Sch_Start_Hour'], kde=True, bins=30)
plt.title('Distribution of Actual Trip Duration')
plt.xlabel('Actual_Trip_End_Minutes')
plt.ylabel('Frequency')
```

```python
plt.subplot(2, 2, 3)
sns.boxplot(x='Operated_Day', y='Actual_Trip_End_Minutes', data=DS)
plt.title('Trip Duration vs. Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Actual Trip Duration (minutes)')
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6], labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
```

```python
plt.subplot(2, 2, 4)
sns.boxplot(x='Route_Name', y='Actual_Trip_End_Minutes', data=DS)
plt.title('Trip Duration vs. Route')
plt.xlabel('Route Name')
plt.ylabel('Actual Trip Duration (minutes)')
plt.xticks(rotation=45, ha='right')
```

```python
# Select features and target variable
X = DS[['Sch_Start_Hour', 'Sch_Start_Minute', 'Operated_Day_Encoded', 'Route_Name_Encoded']]
y = DS['Actual_Trip_End_Minutes']
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
RF_model = RandomForestRegressor()
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 500, num = 10)]
print(n_estimators)
# Initialize and train the model


```

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 500, num = 10)]
# Number of features to consider at every split
max_features = ['log2', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 20, num = 4)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5]
```

```python
rf = RandomForestRegressor()
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = -1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
# Predict and evaluate the model
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

```

```python
print('MAE',mae)
print('RMSE',rmse)
```
